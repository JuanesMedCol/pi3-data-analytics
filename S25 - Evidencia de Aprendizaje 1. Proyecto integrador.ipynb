Cargamos en dataset desde la web con el código que Kaggle proporciona, en esta parte se carga desde Kagglehub, para que no proporcione un ruta de extración del archivo


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

print("Path to dataset files:", path)
```

A continuación se carga en archivo csv desde la ruta proporcionada por kagglehub.

```python
#cargar el dataset anterior
df=pd.read_csv(path+"/marketing_campaign.csv",sep="\t")
```
